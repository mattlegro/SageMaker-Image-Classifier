```python
import sagemaker, boto3, json
from sagemaker.session import Session
from tqdm import tqdm

sagemaker_session = Session()

aws_role = sagemaker_session.get_caller_identity_arn()
aws_region = boto3.Session().region_name
sess = sagemaker.Session()
s3 = boto3.Session().resource('s3')
```

```python
bucket_name = "lets-id-fish"
old_prefix = "species-id-images-augmented/Acanthuridae"
bucket_ref = s3.Bucket(bucket_name)

for object in bucket_ref.objects.filter(Prefix=old_prefix):
    src_key = object.key
    if not src_key.endswith('/'):
        split_key = src_key.split('/')
        filename = split_key[-1]
        sci_name = split_key[-2]
        new_dest_key = f"{old_prefix}/{sci_name}/{filename}"
        copy_source = bucket_name + '/' + src_key
        print(copy_source)
        print(new_dest_key)
```

```python
bucket_name = "lets-id-fish"
old_prefix = "species-id-images-augmented"
bucket_ref = s3.Bucket(bucket_name)

for object in tqdm( bucket_ref.objects.filter(Prefix=old_prefix), desc="Moving Files...", total=60040 ):
    src_key = object.key
    if not src_key.endswith('/'):
        split_key = src_key.split('/')
        filename = split_key[-1]
        sci_name = split_key[-2]
        new_dest_key = f"{old_prefix}/{sci_name}/{filename}"
        copy_source = bucket_name + '/' + src_key
        s3.Object(bucket_name,new_dest_key).copy_from(CopySource=copy_source)
        s3.Object(bucket_name,src_key).delete()
```
